```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```
data = pd.read_csv("homicide.csv", encoding='Latin-1')
data = pd.DataFrame(data)
```

```
data = data.dropna()
```

```
data
```

```
paises = data['Country'].drop(range(19,458,2), errors='ignore',axis=0)
paises = paises.reset_index(drop=True)

rate_homi_1990 = data['2007'].drop(range(18,458,2), errors='ignore',axis=0)
rate_homi_1990=pd.DataFrame(rate_homi_1990)
rate_homi_1990 = rate_homi_1990.reset_index(drop=True)

data_1990 = data["2007"].drop(range(19,458,2), errors='ignore',axis=0)
data_1990 = data_1990.reset_index(drop=True)

plt.pie(data_1990, labels=paises,autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```
aux = pd.merge(paises, data_1990, left_index=True, right_index=True)
homicidios_1990 = pd.merge(aux, rate_homi_1990, left_index=True, right_index=True)
homicidios_1990.to_csv('homicidios_2007.csv')
files.download("homicidios_2007.csv")
```
