```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("homicidios_1990.csv", sep=',', header=0, index_col=0)
```

```
df
```

```
print(df['Homicides'].nlargest(5))
print(df['Homicides'].nsmallest(5))
```

```
paises = df["Country"]
mapear = range(len(paises))


plt.plot(df["Homicides"])
plt.xticks(mapear, paises)
plt.ylabel("Homicidios")
plt.xlabel("Pais")
plt.xlim(42,43)


plt.show()
```

```
rate_homicidios = df["Homicides"]
paises = df["Country"]
plt.title('Tasa de muertes 1990')
plt.pie(rate_homicidios, labels=paises,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```
