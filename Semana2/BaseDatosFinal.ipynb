```
#Creamos los DataFrames
data_lluvia = pd.read_csv("201901010000Lluv.csv", encoding='Latin-1')
data_Tmax = pd.read_csv("201901010000TMax.csv", encoding='Latin-1')
data_Tmed = pd.read_csv("201901010000TMed.csv", encoding='Latin-1')
data_Tmin = pd.read_csv("201901010000TMin.csv", encoding='Latin-1')
```

```
#Queremos llegar a un formato tal que
#LON/LAT/CLAVE/EDO/TMAX/TMED/TMIN

temp = pd.merge(data_Tmax,data_Tmed)
tempe = pd.merge(temp,data_Tmin)
```

```
"""
La data de la lluvia está en mayúsculas y queremos que solo
la letra inicial este en mayúsculas para que sea del mismo 
formato a las tablas de temperatura.

Hacemos una funcion lambda que le pasara cada head de cada
columna como argumento para aplicar el metodo de la clase
string capitalize().

Esto transforma el codigo de por ejemplo: "greeks" a "Greeks"

Lo cual nos funciona para resolver este problema.
"""
data_lluvia.rename(columns=lambda m: m.capitalize())
```

```
#Ahora como estan del mismo formato, se pueden mezclar ambas tablas
Base_Data = pd.merge(tempe, data_lluvia)
```
