```
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```
# Puntos aleatorios alrededor de 0 en una 2-Bola

N = 1500
r = 9

x = np.zeros(N)
y = np.zeros(N)

xe = np.zeros(N)
ye = np.zeros(N)

for i in range(N):
  x0 = np.random.uniform(-r, r)
  y0 = np.random.uniform(-r, r)
  if(x0**2+y0**2 <= r**2):
    x = np.append(x, x0)
    y = np.append(y, y0)
  else:
    xe = np.append(xe,x0)
    ye = np.append(ye, y0)

plt.scatter(x,y, marker=".", c="blue")
plt.scatter(xe,ye,marker=".",c="red")
plt.xlim(-15,15)
plt.ylim(-10,10)
plt.show()
```

```
# Puntos aleatorios alrededor de 0 en una 3-Bola

N = 800
r = 9

x = np.zeros(N)
y = np.zeros(N)
z = np.zeros(N)

xe = np.zeros(N)
ye = np.zeros(N)
ze = np.zeros(N)

for i in range(N):
  x0 = np.random.uniform(-r, r)
  y0 = np.random.uniform(-r, r)
  z0 = np.random.uniform(-r, r)
  m2 = x0**2+y0**2+z0**2
  if( m2 <= r**2):
    x = np.append(x, x0)
    y = np.append(y, y0)
    z = np.append(z, z0)
  else:
    xe = np.append(xe,x0)
    ye = np.append(ye,y0)
    ze = np.append(ze,z0)

#fig1 = plt.scatter(x,y,z, marker=".", c="blue")
#fig2 = plt.scatter(xe,ye,ze, marker=".",c="red")

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(x, y, z, marker='.', c="blue")
ax1.scatter(xe, ye, ze, marker='.', c="red")

plt.xlim(-15,15)
plt.ylim(-10,10)

plt.show()
```
